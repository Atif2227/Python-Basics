## Series Object

#### Series Object is one dimentional labelled object.

```python
import pandas as pd
s1 = pd.Series([1,2,3,4,5])
s1
```

```python
# In the above output 0-1 is teh index number and 1-5 are the value stores at corrsponding indices 
```

```python
type(s1)
```

### Changing Index

```python
import pandas as pd #invoking pandas library is not required every time for a new set of codes.
s1 = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])
s1
```

### Creating series object fron 'Dictionary'

```python
pd.Series({'a':10,'b':20,'c':30})
```

### Changing index position

```python
pd.Series({'a':10,'b':20,'c':30}, index = ['b','c','d','a'])
```

### Extracting Individual Elements

```python
#Extracting Single element

s1 = pd.Series([1,2,3,4,5,6,7,8,9])
s1[3]
```

```python
#Extracting a Sequence of elements

s1[:4] # This will extract first 4 elements, from index 0 to 3.
```

```python
#Extracting Elements from back

s1[-3:] # This will extract the last 3 elements
```

### Basic Maths operation on Series

```python
# Adding a scaler value to Series Elements

s1 + 5
```

```python
# Adding two Series object

s1 = pd.Series([1,2,3,4,5,6,7,8,9])
s2 = pd.Series([10,20,30,40,50,60,70,80,90])
s1 + s2
```

```python
# Multiplication of two Series object

s1 * s2
```

```python
# Division of two Series object

s1/s2
```
