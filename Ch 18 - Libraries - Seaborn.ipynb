# Seaborn

### A library used for visualization and it is build over matplotlib. 

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.shape
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

### Grouping data with 'hue'

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="event")
plt.show()
```

### Addind style

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="event", style="event")
plt.show()
```

### Adding markers

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="event", markers=True, data=fmri)
plt.show()
```

### Seaborn Bar-Plot

```python
import pandas as pd
sns.set(style = "whitegrid")
pokemon = pd.read_csv('pokemon.csv')
pokemon.head()
```

```python
sns.barplot(x="Legendary", y="Speed", data=pokemon)
plt.show()
```

### Sea Born Bar-Plot

```python
sns.barplot(x="Legendary", y="Defense", data=pokemon)
plt.show()
```

```python
sns.barplot(x="Legendary", y="Speed", hue="Generation", data=pokemon)
plt.show()
```

### Changing bar colors with 'Palette'

```python
sns.barplot(x="Legendary", y="Defense", data=pokemon, palette='Blues_d')
plt.show()
```

```python
sns.barplot(x="Legendary", y="Defense", data=pokemon, palette='rocket')
plt.show()
```

```python
sns.barplot(x="Legendary", y="Defense", data=pokemon, palette='rainbow_r')
plt.show()
```

### Fixing Bar Colours 

```python
sns.barplot(x="Legendary", y="Defense", color="lightgreen", data=pokemon)
plt.show()
```

### Seaborn Scatter Plot

```python
iris=pd.read_csv('iris.csv')
iris.head()
```

```python
sns.scatterplot(x="SepalLengthCm", y='PetalLengthCm', color= "green",data=iris)
plt.show()
```

### Adding Hue & Style to scatter-plot

```python
sns.scatterplot(x="SepalLengthCm",y="PetalWidthCm", hue="Species", style="Species", data=iris)
plt.show()
```

```python
sns.scatterplot(x="SepalLengthCm",y="PetalWidthCm", hue="PetalWidthCm", data=iris)
plt.show()
```

### Seaborn Histogram/Distribution Curve

```python
diamonds = pd.read_csv("diamonds.csv")
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
 sns.histplot(diamonds['price'])
plt.show()
```

```python
sns.displot(diamonds['price'], color='green')
plt.show()
```

```python
sns.kdeplot(diamonds['price'], color='red') # kdeplot only gives the distribution curve and not the histogram.
plt.show()
```

```python
sns.displot(diamonds['price'], color='green', bins=10)
plt.show()
```

```python
sns.distplot(diamonds['price'], color='green', vertical=True)
plt.show()
```

```python
sns.displot(y='price', color= 'orange', data=diamonds)
plt.show()
```

### Seaborn Joint-Plot (Scatter plot + HIstogram)

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.jointplot(x="SepalLengthCm", y="PetalLengthCm", color='green', data=iris)
plt.show()
```

#### Adding regresion line in joitplot

```python
sns.jointplot(x="SepalLengthCm", y="PetalLengthCm", color='green', kind= "reg", data=iris)
plt.show()
```

### Seaborn Box-plot

```python
diamonds = pd.read_csv('diamonds.csv')
diamonds.head()
```

```python
sns.boxplot(x='cut', y= 'price', data=diamonds)
plt.show()
```

```python
chur = pd.read_csv('churn.csv')
chur.head()
```

```python
sns.boxplot(x='InternetService', y='MonthlyCharges', data=churn)
plt.show()
```

#### Paletts

```python
sns.boxplot(x='InternetService', y='MonthlyCharges', palette = 'Set1', data=churn)
plt.show()
```

#### Assignig colour

```python
sns.boxplot(x='InternetService', y='MonthlyCharges', color='blue', data=churn)
plt.show()
```

#### Line width - increase or decrease the boundary line of box-plot

```python
sns.boxplot(x='InternetService', y='MonthlyCharges', color='orange', linewidth=4, data=churn)
plt.show()
```

#### change Order

```python
sns.boxplot(x='InternetService', y='MonthlyCharges', color='orange', data=churn, order=["No","DSL","Fiber optic"])
plt.show()
```

#### Addinh Hue

```python
sns.boxplot(x='InternetService', y='MonthlyCharges', hue= 'PaymentMethod', data=churn)
plt.show()
```
