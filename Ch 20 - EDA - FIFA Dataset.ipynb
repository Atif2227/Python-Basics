```python
#Loading required libraries

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fifa = pd.read_csv('players_20.csv')
fifa.head()
```

```python
#Top 10 countries(nationality) with maximum number of players.
fifa['nationality'].value_counts()[0:10]
```

```python
#Extract top 10 countries(nationality) with maximum number of players.
fifa['nationality'].value_counts()[0:10].keys()
```

```python
#Extract the number of players for top 10 countries(nationality) with maximum number of players.
fifa['nationality'].value_counts()[0:10]
```

```python
#make a bar plot for top 5 countries with maximum number of same nationality.
plt.figure(figsize=(7,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()), list(fifa['nationality'].value_counts()[0:5]), 
        color=['red','green','blue','orange','pink'])
plt.show()
```

```python
#Salary of players

salary = fifa[['short_name','wage_eur']]
```

```python
salary.head()
```

```python
#Top 5 players with maximum salaries

salarytop=salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
salarytop.head(5)
```

```python
#make a bar-plot for above result

plt.figure(figsize=(8,5))
plt.bar(list(salarytop['short_name'])[0:5],
        list(salarytop['wage_eur'])[0:5], color=['r','g','b','yellow','skyblue'])
plt.show()
```
